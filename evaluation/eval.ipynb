# Utils

```python
import matplotlib.pyplot as plt

def plot_dist_ft(df_full_col, df_clean_col, score_type):

    plt.figure(figsize=(10, 6))
    # original model
    plt.hist(df_full_col, bins=20, alpha=0.5, color='skyblue', edgecolor='black', label='BART-swipe-full')
    # fine-tuned model
    plt.hist(df_clean_col, bins=20, alpha=0.5, color='orange', edgecolor='black', label='BART-swipe-clean')
    plt.title(f'{score_type} Scores', fontsize=16, fontweight='bold')
    plt.xlabel(f'{score_type} Score', fontsize=14)
    plt.ylabel('Number of Texts', fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend()
    
    plt.show()


def plot_dist_orig(df_col_full, df_col_clean, score_type):

    plt.figure(figsize=(10, 6))
    # original model
    plt.hist(df_col_full, bins=20, alpha=0.5, color='skyblue', edgecolor='black', label='BART-large-swipe')
    # fine-tuned model
    plt.hist(df_col_clean, bins=20, alpha=0.5, color='orange', edgecolor='black', label='BART-large-swipe-clean')
    plt.title(f'{score_type} Scores', fontsize=16, fontweight='bold')
    plt.xlabel(f'{score_type} Score', fontsize=14)
    plt.ylabel('Number of Texts', fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend()
    
    plt.show()

def plot_dist_orig_ft_3(df_orig_full_col, df_orig_clean_col, df_ft_clean_col, score_type):

    plt.figure(figsize=(10, 6))
    # original model
    plt.hist(df_orig_full_col, bins=20, alpha=0.5, color='skyblue', edgecolor='black', label='BART-large-swipe-full')
    # orig
    plt.hist(df_orig_clean_col, bins=20, alpha=0.5, color='orange', edgecolor='black', label='BART-large-swipe-clean')
    # fine-tuned
    plt.hist(df_ft_clean_col, bins=20, alpha=0.5, color='red', edgecolor='black', label='Fine-tuned BART-swipe-clean')
    
    plt.title(f'{score_type} Scores', fontsize=16, fontweight='bold')
    plt.xlabel(f'{score_type} Score', fontsize=14)
    plt.ylabel('Number of Texts', fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend()
    
    plt.show()
```
